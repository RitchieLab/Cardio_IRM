# load packages

```python
import pandas as pd
```

# make dataframe

```python
filepath_list = []
for chr in list(range(1, 23)):
    filepath = 'ukb21007_c' + str(chr) + '_b0_v1.topmed_imputed.cad_afib_pgs_vars'
    filepath_list.append(filepath)
print(len(filepath_list))
```

```python
samplesheet = pd.DataFrame({'sampleset' : (['UKBB'] * 22),
                           'path_prefix' : filepath_list,
                           'chrom' : list(range(1, 23)),
                           'format' : (['pfile'] * 22)})
print(samplesheet.shape)
samplesheet.head()
```

# export

```python
samplesheet.to_csv('UKBB.CAD.AFIB.PGSC_CALC.samplesheet.csv', index = None)
```
