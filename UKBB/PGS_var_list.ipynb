# load packages

```python
import pandas as pd
```

# read in input files

```python
cad_pgs = pd.read_csv('PGS005092_hmPOS_GRCh38.txt.gz', sep = '\t', comment = '#')
print(cad_pgs.shape)
cad_pgs.head()
```

```python
afib_pgs = pd.read_csv('PGS005072_hmPOS_GRCh38.txt.gz', sep = '\t', comment = '#')
print(afib_pgs.shape)
afib_pgs.head()
```

# subset

```python
cad_pgs_sub = cad_pgs[['chr_name', 'chr_position']]
cad_pgs_sub['chr_position2'] = cad_pgs_sub['chr_position']
cad_pgs_sub.head()
```

```python
afib_pgs_sub = afib_pgs[['chr_name', 'chr_position']]
afib_pgs_sub['chr_position2'] = afib_pgs_sub['chr_position']
afib_pgs_sub.head()
```

# cat and drop duplicates

```python
all_vars = pd.concat([cad_pgs_sub, afib_pgs_sub], axis = 0).drop_duplicates().sort_values(by = ['chr_name', 'chr_position'])
print(len(all_vars.index))
all_vars.head()
```

# export

```python
all_vars.to_csv('CAD.PGS005092.AFIB.PGS005072.var_list.txt',
                sep = '\t',
                header = None,
                index = None)
```
