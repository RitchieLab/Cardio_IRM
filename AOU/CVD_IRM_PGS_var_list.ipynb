# read in input files

```python
cad_pgs = pd.read_csv('PGS005112_hmPOS_GRCh38.txt.gz', sep = '\t', comment = '#')
print(cad_pgs.shape)
cad_pgs.head()
```

```python
afib_pgs = pd.read_csv('PGS005072_hmPOS_GRCh38.txt.gz', sep = '\t', comment = '#')
print(afib_pgs.shape)
afib_pgs.head()
```

```python
hf_pgs = pd.read_csv('PGS005097_hmPOS_GRCh38.txt.gz', sep = '\t', comment = '#')
print(hf_pgs.shape)
hf_pgs.head()
```

# subset and rename

```python
cad_pgs_sub = cad_pgs[['hm_chr', 'hm_pos']]
cad_pgs_sub['chr_position2'] = cad_pgs_sub['hm_chr']
cad_pgs_sub = cad_pgs_sub.rename(columns = {'hm_chr' : 'chr_name', 'hm_pos' : 'chr_position'})
cad_pgs_sub.head()
```

```python
afib_pgs_sub = afib_pgs[['chr_name', 'chr_position']]
afib_pgs_sub['chr_position2'] = afib_pgs_sub['chr_position']
afib_pgs_sub.head()
```

```python
hf_pgs_sub = hf_pgs[['chr_name', 'chr_position']]
hf_pgs_sub['chr_position2'] = hf_pgs_sub['chr_position']
hf_pgs_sub.head()
```

# cat and drop duplicates

```python
all_vars = pd.concat([cad_pgs_sub, afib_pgs_sub], axis = 0).drop_duplicates().sort_values(by = ['chr_name', 'chr_position'])
print(len(all_vars.index))
all_vars.head()
```

# export

```python
all_vars.to_csv('CAD.PGS005112.AFIB.PGS005072.var_list.txt',
                sep = '\t',
                header = None,
                index = None)
```

```python
hf_pgs_sub.to_csv('HF.PGS005097.var_list.txt',
                  sep = '\t',
                  header = None,
                  index = None)
```
